# =====필요한 기록만 추출======

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from matplotlib import rc, font_manager
rc('font', family='NanumSquareR')
```

```python
os.chdir('C:/Users/goddo/Desktop/dacon/data')
regular=pd.read_csv("Regular_Season_Batter.csv")
daybyday=pd.read_csv("Regular_Season_Batter_Day_by_Day.csv")
pre=pd.read_csv("Pre_Season_Batter.csv")
sub=pd.read_csv("submission.csv")

pd.set_option('display.max_columns', 500)
```

```python
# submission에 없는 타자 빼자
regularNamelist=list(regular["batter_name"].unique())
subNameslist= list(set(sub['batter_name']))
namelist=[]
for i in regularNamelist:
    if i in subNameslist:
        namelist+=[i]

delrow=[]
for i in range(len(regular)):
    if regular["batter_name"][i] in namelist:
        continue
    else:
        delrow+=[i]
regular = regular.drop(delrow,0)
```

```python
# regular 타석 100이상인 경우로 한정시킨다.
regular = regular.loc[(regular['AB'] >= 100)]

# regular column중 'OPS'와 관련 없는 'team', 'E', 'height/weight', 'year_born', 'career', ''starting_salary', 'SB', 'CS', 'RBI', 'position', 'GDP'] 삭제
regular = regular.drop(['team','E','height/weight','year_born','career','starting_salary', 'SB', 'CS', 'RBI', 'position', 'GDP'], axis=1)
# BABIP, 1B 칼럼 추가 (분모에 희생플라이(SF)를 추가하여야 하나, 수치가 적고 데이터 수집 시간상 생략하였음)
regular['BABIP'] = (regular['H']-regular['HR']) / (regular['AB']-regular['SO']-regular['HR'])
regular['1B'] = regular['H'] - regular['2B'] - regular['3B'] - regular['HR']
```

```python
# 1B 칼럼 'H'와 '2B' 사이로 이동
record = ['batter_id','batter_name','year','avg','G','AB','R','H','1B','2B','3B','HR','TB','BB','HBP','SO','SLG','OBP','OPS','BABIP']
regular = regular.reindex(columns=record)
```

# LAG BABIP 비교 상관계수: 0.351

```python
var = 'BABIP'
x = []
y = []


for name in regular['batter_name'].unique():
    a = regular.loc[regular['batter_name']==name,]
    k = []
    for i in a['year'].unique():
        if (a['year']==i+1).sum()>0:
            k.append(i)
    for i in k:
        x.append(a.loc[a['year']==i, var].iloc[0])
        y.append(a.loc[a['year']==i+1, var].iloc[0])
plt.scatter(x,y)
plt.title(var)
plt.xlabel('(y)year BABIP')
plt.ylabel('(y+1)year BABIP')
print(pd.Series(x).corr(pd.Series(y)))
```
